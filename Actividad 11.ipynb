```python
#Analitica de datos 
#Arreglo de datos de 2 dimensiones 

#Importando las bibliotecas necesarias
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

#Generando datos aleatorios 
np.random.seed(42) #para la reproductibidad
data = np.random.randn(100) #100 puntos de datos aleatorios de una sistribucion normal 

#Creando un DataFrame con pandas
df = pd.DataFrame(data, columns=['Valores'])

#Calculando estadisticas descriptivas con scipy y pandas
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
std_dev = np.std(data)

#Imprimir estadisticas descriptivas
print(f"Media: {mean}")
print(f"Mediana: {median}")
print(f"Moda: {mode.mode[0]} (aparece {mode.count[0]} veces)")
print(f"Desviacion estandar: {std_dev}")

#Visualizando los datos con matplotlib
plt.figure(figsize =(10,6))
plt.hist(data, bins = 20, color='blue', alpha = 0.7)
plt.axvline(mean,color='red', linestyle='dashed',linewidth=1, label='Media')
plt.axvline(median, color='yellow', linestyle='dotted',linewidth=1, label='Mediana')
plt.axvline(mode.mode[0], color='green', linestyle='dotted',linewidth=1, label='Moda')
plt.title('Distribucion de los datos')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
      
```

Arreglo de datos, de tres dimensiones

```python
#Importar las bibliotecas necesarias:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Generar datos trididmensionales:
#Vamos a asumir que quieras una matriz 10x10x10. Esto se puede adaptar segun tus necesidades 

x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)
z = np.linspace(0, 10, 10)

X, Y, Z = np.meshgrid(x, y, z)
Values = np.random.rand(10, 10, 10) # Este sera un valor aleatorio para cada combinacion de x, y, s

#Crear un DaraFrame de pandas con los datos:
df = pd.DataFrame({
    'X' : X.flatten(),
    'Y' : Y.flatten(),
    'Z' : Z.flatten(),
    'Values' : Values.flatten()
})

#Visualizar los datos en 3D usando matplotlib:

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(df['X'], df['Y'], df['Z'], c=df['Values'], cmap='viridis', s=50)
plt.colorbar(sc)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Data visualizacion con matplotlib')

plt.show()
```
